```python
import yaml
import os
```

```python
# The base configuration for your YAML files
base_config = {
    'N': 100000,
    'd': 300,
    'output_dir': 'output',
    'trial_num': 100,
    'noise_level_list': [0, 0.01, 0.05, 0.1],
    'hydra': {
        'hydra_logging': {
            'level': 'CRITICAL'
        },
        'job_logging': {
            'level': 'CRITICAL'
        },
        'run': {
            'dir': './outputs',
        }
    },
}

# The list of values for your `true_sparsity_list` parameter
values = [2, 5, 10, 20, 100, 200, 2000]

# The path to the directory where you want to save your YAML files
output_dir = "./Configs/"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Generate a YAML file for each pair of values
for i in range(0, len(values), 2):
    # Copy the base configuration
    config = base_config.copy()

    # Set the `true_sparsity_list` parameter to a list containing the current pair of values
    config['true_sparsity_list'] = values[i:i+2]

    # Define the output file name
    output_file = os.path.join(output_dir, f"sparse_level_{values[i]}{'_'+str(values[i+1]) if i + 1 < len(values) else ''}.yaml")

    # Write the configuration to the output file
    with open(output_file, 'w') as f:
        yaml.dump(config, f)
```
